```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
x=np.array([12,23,34,45,56,67,78,89,90,10])
y=np.array([77,22,55,66,44,88,95,45,89,44])
z=np.array([89,64,55,34,54,79,34,34,88,78])
```

line plot

ms = marker size<br>
mfc= marker font colour<br>
ls= line style<br>



```python
plt.title('line plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.plot(x,y, marker='o',color='blue',mfc='r',ls='--',ms=7);
plt.grid()
```

```python
plt.title('line plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.plot(x,z, marker='*',color='r',mfc='blue',ls='--',ms=7);
plt.grid()
```

```python
plt.title('line plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.plot(x,z, marker='*',color='r',mfc='blue',ls='--',ms=7);
plt.plot(x,y, marker='o',color='blue',mfc='r',ls='-.',ms=7);
plt.plot(z,y, marker='H',color='blue',mfc='g',ls=':',ms=7);
plt.grid()
```

Subplot


```python
plt.subplot(2,2,1)
#plt.figure(figsize=(10,10))
plt.title('line plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.plot(x,y, marker='o',color='blue',mfc='r',ls='-.',ms=7);
plt.grid()
plt.subplot(2,2,2)
plt.title('line plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.plot(x,z, marker='*',color='r',mfc='blue',ls=':',ms=7);
plt.grid()
```

Bar Graph

```python
students=[55,86,77,61,41]
courses=['Java','C','python','c++','Dart']
plt.bar(courses,students,color='black',edgecolor='red',width=0.6,linewidth=1.5,alpha=0.8);
plt.title('Bar Graph')
plt.xlabel('Course')
plt.ylabel('Students')
```

Scatter Plot

```python
plt.scatter(x,y,marker='o')
plt.grid()
```

```python
plt.scatter(y,z,marker='H')
plt.grid()
```

Pie Chart

```python
plt.title('Pie Chart')
plt.pie(students,labels=courses,radius=0.8,labeldistance=1.2,shadow=True,autopct='%.2f%%0',startangle=30,explode=[0,0,0,0.1,0])
plt.grid()
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,4,1)
plt.bar(courses,students,color='black',edgecolor='red',width=0.6,linewidth=1.5,alpha=0.8);
plt.title('Bar Graph')
plt.xlabel('Course')
plt.ylabel('Students')
plt.subplot(1,4,2)
plt.title('Pie Chart')
plt.pie(students,labels=courses,radius=0.8,labeldistance=1.2,shadow=True,autopct='%.2f%%0',startangle=30,explode=[0,0,0,0.1,0])
plt.grid()
plt.subplot(1,4,3)
#plt.figure(figsize=(10,10))
plt.title('line plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.plot(x,y, marker='o',color='blue',mfc='r',ls='-.',ms=7);
plt.grid()
plt.subplot(1,4,4)
plt.title('line plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.plot(x,z, marker='*',color='r',mfc='blue',ls=':',ms=7);
plt.grid()
plt.subplots_adjust(wspace=0.6)
plt.savefig('my plot')
```
