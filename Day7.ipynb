```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.array([67, 54, 17, 83, 14, 59, 63, 77])
```

```python
#plotting x values  with the array [0, 1, 2, 3, 4, 5, 6, 7]
plt.scatter(x, range(8))
plt.scatter(x.mean(), 3.5, color = 'red')
```

```python
#manually finding std
std = np.sqrt( np.sum( (x-x.mean())**2 )/len(x) )
std
```

```python
#checking with inbuilt std function
x.std()
```

```python
#function to make std. dev 1
def standardize(x):
    val = (x-x.min())/x.std()
    return val

x_new = standardize(x)
x_new
```

```python
df=pd.read_csv('/content/wisc_bc_data.csv')
```

```python
df
```

```python
#separate the input and output variable
x=df.drop(['id','diagnosis'],axis=1)
y=df['diagnosis']
```

```python
x.shape
```

```python
from sklearn.feature_selection import SelectKBest, chi2
```

```python
skf=SelectKBest(score_func=chi2, k=5)
```

```python
skf.fit(x,y)
```

```python
skf.get_support()
```

```python
x_new=x.iloc[:,skf.get_support()]
x_new
```
