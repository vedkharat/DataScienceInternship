Data scaling

Normalization

```python
x=np.array([12,32,59,78,56,34,75,45])
```

```python
def normalize(a):
  norm = (a-a.min())/(a.max()-a.min())
  return norm
```

```python
normalize(x)
```

```python
nx=normalize(x)
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.title('Actual Data')
plt.plot(x)
plt.subplot(1,2,2)
plt.title('Normalize Data')
plt.plot(nx)
```

```python
df=pd.read_csv('/content/Social_Network_Ads.csv')
df
```

```python
x=df[['Age','EstimatedSalary']]
y=df['Purchased']
```

```python
plt.xlabel('age')
plt.ylabel('salary')
plt.scatter(x['Age'],x['EstimatedSalary'],c=y)
plt.grid()
```

```python
x.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
x_scaled=scaler.fit_transform(x)
```

```python
type(x_scaled)
```

```python
x_new=pd.DataFrame(x_scaled,columns=x.columns)
```

```python
x_new
```

```python
x_new.describe()
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.title('Actual Data')
plt.xlabel('age')
plt.ylabel('salary')
plt.scatter(x['Age'],x['EstimatedSalary'],c=y)
plt.grid()
plt.subplot(1,2,2)
plt.title('Normalize Data')
plt.xlabel('age')
plt.ylabel('salary')
plt.scatter(x_new['Age'],x_new['EstimatedSalary'],c=y)
plt.grid()
```
