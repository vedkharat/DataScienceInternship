```python
df=pd.read_csv('/content/student.csv')
df.head(3)
```

```python
x=['M','M','M','M','F','M','F','F','M','M']
df['gender']=x
df
```

```python
ndf=pd.DataFrame({
    'roll': range(1,11),
    'cgpa': df['marks']/10+1.00,
    'gender': ['M','M','M','M','F','M','F','F','M','M']
})
```

```python
ndf
```

```python
df.merge(ndf,on='roll')
```
